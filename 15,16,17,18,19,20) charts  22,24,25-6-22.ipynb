```python
#BAR CHART:
import matplotlib.pyplot as plt
import numpy as np
country=('USA','INDIA','UK','FRANCE')
GDP_Per_Capita=(20000,23000,25000,26000)
plt.bar(country,GDP_Per_Capita)
plt.title('Country vs GDP per capita')
plt.xlabel('country')
plt.ylabel('GDP Per Capita')
plt.show()
```

```python
#SUB PLOT(ROWS,COLUMNS,SEQUENCE)
import matplotlib.pyplot as plt
import numpy as np
#plot1
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
plt.subplot(1,2,1)
plt.plot(x,y)
#plot2
x=np.array([1,2,3,4,5])
plt.subplot(1,2,2)
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([0,2,5,7,11])
plt.subplot(2,1,1)
plt.plot(x,y)
#plot2
x=np.array([10,50,30])
plt.subplot(2,1,2)
plt.pie(x)
plt.show()
```

```python
#plot1
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
plt.subplot(2,2,1)
plt.plot(x,y)
#plot2
x=np.array([1,2,3,4,5])
plt.subplot(2,2,2)
plt.pie(x)
#plot3
x=np.array([1,2,3,4,5])
y=np.array([100,20,50,95,80])
plt.subplot(2,2,3)
plt.plot(x,y)
#plot4
x=np.array([10,60,30])
plt.subplot(2,2,4)
plt.pie(x)
plt.show()
```
