```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
xpoints=a["Name"].tolist()
ypoints=a["salary"].tolist()
zpoints=a["Age"].tolist()
apoints=a["profit"].tolist()
plt.plot(xpoints,ypoints,marker="s",linewidth=3,mec="red",color="blue")
plt.plot(xpoints,zpoints,marker="o",linewidth=2,mec="red",color="green")
plt.plot(xpoints,apoints,marker="*",linewidth=3,mec="red",color="orange")
plt.show()
```
